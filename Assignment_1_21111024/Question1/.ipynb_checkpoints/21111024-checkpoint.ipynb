## KNN algorithm to predict the prices 

```python
class KNN:
    def __init__(self, k):
        self.k = k
    
    def fit(self, X_train, y_train):
        self.X_train = X_train
        self.y_train = y_train
    
#     function to find euclidian distance
    def find_distance(self, point1, point2):
        dist = np.linalg.norm(point1 - point2)
        return dist
    
    def predict(self, X_test):
        result = []
        for i in range(len(X_test)):
            dist = []
            nearest = []
#           find euclidian distance from all training points
            for j in range(len(self.X_train)):
                dist.append([self.find_distance(X_test[i],self.X_train[j]),j])
#           sorting distances
            dist.sort()
#           finding k neareset distances
            dist = dist[:self.k]
            for a, b in dist:
                nearest.append(self.y_train[b])
#           finding average of all nearest items
            result.append(np.mean(nearest))
        return result
    
#     function to find rmse value
    def score(self, actual, predicted):
        return np.sqrt(((predicted-actual)**2).mean())
```

```python
def knn_train_test(train, test, features, target,k):
    X_train = train[features].values
    y_train = train[target].values
    X_test = test[features].values
    y_test = test[target].values
    knn = KNN(k)
    knn.fit(X_train, y_train)
    prediction = knn.predict(X_test)
    rmse = knn.score(y_test,prediction)
    return rmse
```

```python
# splitting data in to train and test set
train=data.sample(frac=0.8,random_state=100)
test=data.drop(train.index).sample(frac=1.0)
train_cols = data.columns.drop('price')
rmse = knn_train_test(train, test, train_cols.values, 'price', 1)
print("Root mean square error is",rmse)
```

#### Root means square based on our prediction is 3788.21

## Cross validation to find the optimal value of K 

```python
def kFold(data,k, neighbors):
    error_values=[]
    train_cols = data.columns.drop('price')
    data = data.sample(frac=1.0,random_state=100)
#   creating 10 folds
    folds=np.array_split(data, k)
    for j in range(k):
        total = folds[:j]+folds[j+1:]
        train = pd.concat(total)
        test = folds[j]
#       finding rmse value for each fold
        rmse=knn_train_test(train,test,train_cols,'price',neighbors)
        error_values.append(rmse)
        
#   returning mean of all rmse values
    return np.mean(error_values)
```

```python
def findOptimalK(data):
    k_values = [i for i in range(1,30)]
    error_values = []
    for i in k_values:
#         folding data 10 times
        error=kFold(data, 10, i)
        error_values.append(error)
    plt.plot(k_values,error_values)
    plt.xlabel("no of neighbors")
    plt.ylabel("mean rmse value")
    plt.show()
    print("minimum error is ",min(error_values))
    return error_values.index(min(error_values))+1
```

```python
k = findOptimalK(data)
print("optimal value of k is",k)
```

#### Minimum RMSE  is 2649.89
#### Optimal value of K is 2
